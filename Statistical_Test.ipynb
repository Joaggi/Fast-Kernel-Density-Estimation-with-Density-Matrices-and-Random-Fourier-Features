```python
# Friedman test
import numpy as np
from numpy.random import seed
from numpy.random import randn
from scipy.stats import friedmanchisquare
# seed the random number generator
seed(1)
# generate three independent samples
ARC = [52400,103000,24700,49200,56000,4330] 
TWO_MIXT= [42400,68000,35900,54800,62000,3500]  
POT1= [46100,76000,22100,39900,52400,3640] 
POT2= [28800,70000,4920,29400,36700,3630] 
POT3= [30200,72000,19900,41100,48900,3280] 
POT4 =  [30700,68000,54700,54800,64000,3700] 
# compare samples
stat, p = friedmanchisquare(ARC,TWO_MIXT,POT1,POT2,POT3,POT4)
print('Statistics=%.3f, p=%.3f' % (stat, p))
# interpret
alpha = 0.05
if p > alpha:
	print('Same distributions (fail to reject H0)')
else:
	print('Different distributions (reject H0)') 
```
